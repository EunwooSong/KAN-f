```python
#data  JSONL document로 변경
import json

# 1번 데이터 파일 로드
with open("ko_converted_data_modified.json", "r", encoding="utf-8") as f:
    original_data = json.load(f)

# 변환된 데이터를 저장할 파일 생성
with open("prompt_data.jsonl", "w", encoding="utf-8") as f:
    # 각 원본 데이터에 대해 변환된 데이터를 생성하여 파일에 작성
    for item in original_data:
        origin_text = item["en_origin"]
        prompt_text = item["ko_converted"]
        completion = {
            "prompt": prompt_text,
            "completion": origin_text
        }
        # 변환된 데이터를 파일에 작성 (한 줄씩 작성)
        f.write(json.dumps(completion, ensure_ascii=False) + "\n")

```

```python
#데이터 1/n으로 쪼개기
import json
import random

# 원본 데이터 파일 로드
with open("prompt_data_prepared.jsonl", "r", encoding="utf-8") as f:
    original_data = [json.loads(line) for line in f]

# 데이터 크기를 1/n로 줄이기 (여기서는 n=5으로 가정)
n = 3*2*2*2*2
sampled_data = random.sample(original_data, len(original_data) // n)

# 샘플링된 데이터 저장
with open("prompt_data_prepared_sample.jsonl", "w", encoding="utf-8") as f:
    for item in sampled_data:
        f.write(json.dumps(item, ensure_ascii=False) + "\n")
```

```python
model_engine = "YOUR_ID" # "Davinci" 모델 엔진 ID
prompt = "오늘 서울시 강남구의 날씨가 어때?" # 모델에 입력할 프롬프트

# API 호출
completions = openai.Completion.create(
    engine=model_engine,
    prompt=prompt,
    max_tokens=2048
)

# 결과 출력
for choice in completions.choices:
    print(choice.text)
```
